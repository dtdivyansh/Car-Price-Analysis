## 4) Splitting of data into training and test dataset

### NOTE: Since we have to predict two columns that are 'GT Compressor decay state coefficient' and 'GT Turbine decay state coefficient', we'll work on them simultaneously for better understanding of our models and avoid using different models again and again

```python
x = df.iloc[:,0:17]
y1 = df['GT Compressor decay state coefficient.']
y2 = df['GT Turbine decay state coefficient.']
```

```python
from sklearn.model_selection import train_test_split

# FOR GT Compressor decay state coefficient PREDICTION
x_t1,x_test1,y_t1,y_test1 = train_test_split(x,y1,test_size = 0.2,shuffle = True)

# FOR GT Turbine decay state coefficient PREDICTION
x_t2,x_test2,y_t2,y_test2 = train_test_split(x,y2,test_size = 0.2,shuffle = True)
```

## 5) Importing different machine learning models from Sklearn

```python
from sklearn import linear_model

model1 = linear_model.SGDRegressor(max_iter = 1000, loss='squared_loss')
model2 = linear_model.LinearRegression()
model3 = linear_model.Lasso(alpha=0.2)
model4 = linear_model.Ridge(alpha=0.35)
```

## 6) Training/Fitting our model

```python
# TRAINING FOR GT Compressor decay state coefficient PREDICTION

model1.fit(x_t1,y_t1)

model2.fit(x_t1,y_t1)

model3.fit(x_t1,y_t1)

model4.fit(x_t1,y_t1)

print('Training Complete')
```

### Now our all 4 models are trained, let's check there score  !!

```python
print( 'SGD score: ', model1.score(x_t1,y_t1) )
print( 'Linear regression score: ', model2.score(x_t1,y_t1) )
print( 'Lasso score: ', model3.score(x_t1,y_t1) )
print( 'Ridge score: ', model4.score(x_t1,y_t1) )
```

### Clearly, Stochastic gradient descent and Lasso models turned out to be very poor for our data.

## 7) Checking our models on test dataset

```python
LR_acc = model2.score(x_test1,y_test1)
Ridge_acc = model4.score(x_test1,y_test1)

print('Linear regression test accuracy: ',LR_acc)
print('Ridge test accuracy: ',Ridge_acc)
```

### *This is literally amazing how Linear Regression aced into this model training. No doubt it is the most used regression model.*

## 8) Importing yellowbrick for Visualizing our model perfomance and feature importance for better feature selection while deploying real world model

### A- For Linear Regression

```python
viz = FeatureImportances(model2)

viz.fit(x_t1, y_t1)
viz.show()
```

### B-  FOR RIDGE REGRESSION

```python
viz2 = FeatureImportances(model4)
viz2.fit(x_t1, y_t1)
viz2.show()
```

```python
curve2 = LearningCurve(model4, scoring='r2')
curve2.fit(x_t1, y_t1)
curve2.show()
```

# Curve shows that validation score kept increasing with training instances and Bar graph shows "Level Position", "Fuel Flow", "GT Turbine decay state coefficient" and "HP turbine exit pressure" have quite good impact on model training

### WE CAN USE ONLY THESE FEATURES FOR FITTING THIS MODEL. THIS IS CALLED FEATURE SELECTION.

## # *NOW PREDICTING COLUMN  "GT Turbine decay state coefficient"*

Note: Here we will focus on minimizing RMSE also.

```python
# Linear regression model

model5 = linear_model.LinearRegression()

model5.fit(x_t2,y_t2)
print( 'Linear regression score: ', model5.score(x_t2,y_t2) )

LR_acc = model5.score(x_test2,y_test2)  # Checking Test Accuracy
print('Linear regression test accuracy: ',LR_acc)

y_pred_linear = model5.predict(x_test2)  # Getting Predicted Values

viz3 = FeatureImportances(model5) # Feature Importance

viz3.fit(x_t2, y_t2)
viz3.show()
```

```python
# Ridge Model

model6 = linear_model.Ridge(alpha=0.35)

model6.fit(x_t2,y_t2)
print( 'Ridge Model score: ', model6.score(x_t2,y_t2) )

Ridge_acc = model6.score(x_test2,y_test2) # Checking Test Accuracy
print('Ridge test accuracy: ',Ridge_acc)

y_pred_ridge = model6.predict(x_test2)  # Getting Predicted Values

viz4 = FeatureImportances(model6) # Feature Importance

viz4.fit(x_t2, y_t2)
viz4.show()
```

# Here features   "Fuel Flow", "GT Compressor decay state coefficient", "GT Compressor outlet air pressure" and "HP turbine exit pressure"  are most important features for this model.

### WE CAN USE ONLY THESE FEATURES FOR FITTING THIS MODEL. THIS IS CALLED FEATURE SELECTION.

## *9) Checking RMSE loss function*

```python
#Importing Metrics

from sklearn import metrics

loss_linear = np.sqrt( metrics.mean_squared_error(y_test2,y_pred_linear) )

loss_ridge = np.sqrt( metrics.mean_squared_error(y_test2,y_pred_ridge) )

print("RMSE Loss Linear: ",loss_linear)
print("RMSE Loss Ridge: ",loss_ridge)
```

## Fortunately we are getting very less RMSE loss value for both the models which is really great.

## *Although, we can try to minimize the loss by selecting important features mentioned above. We need not to do that here because we are getting good results.*

```python
curve3 = LearningCurve(model5, scoring='r2')
curve3.fit(x_t2, y_t2)
curve3.show()
```

```python
curve4 = LearningCurve(model6, scoring='r2')
curve4.fit(x_t2, y_t2)
curve4.show()
```

## Clearly we got satisfactory scores with very less loss here as training and cross-validation curves overlapped each other very soon.
